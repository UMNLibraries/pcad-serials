# PCAD notebook 4

This notebook creates a dataframe of item-level enumeration and chronology data from an Alma-published MARC file, and filters out items for libraries and locations not included in the current PCAD run.

Required files/inputs:

- MARC file produced by Alma publishing job `pcad p2e physical serials with items loc data`.
- PCAD libraries and locations Excel file. 

Outputs:

- `.pkl` file of item data with items belonging to non-included libraries/locations removed.

```python
import pymarc
import pandas as pd
from datetime import date
today = str(date.today()).replace('-','')
```

```python
def extract_marc_info(file):
    records = []
    reader = pymarc.MARCReader(file)
    for rec in reader:
        field_001 = rec.get_fields('001')
        field_954 = rec.get_fields('954')
        for f001 in field_001:
            mmsid = str(f001).lstrip('=01')
        for f954 in field_954:
            la_954 = f954.get_subfields('a')
            a_954 = la_954[0] if len(la_954) > 0 else ''
            lb_954 = f954.get_subfields('b')
            b_954 = lb_954[0] if len(lb_954) > 0 else ''
            lc_954 = f954.get_subfields('c')
            c_954 = lc_954[0] if len(lc_954) > 0 else ''
            ld_954 = f954.get_subfields('d')
            d_954 = ld_954[0] if len(ld_954) > 0 else ''
            le_954 = f954.get_subfields('e')
            e_954 = le_954[0] if len(le_954) > 0 else ''
            lf_954 = f954.get_subfields('f')
            f_954 = lf_954[0] if len(lf_954) > 0 else ''
            lg_954 = f954.get_subfields('g')
            g_954 = lg_954[0] if len(lg_954) > 0 else ''
            lh_954 = f954.get_subfields('h')
            h_954 = lh_954[0] if len(lh_954) > 0 else ''
            li_954 = f954.get_subfields('i')
            i_954 = li_954[0] if len(li_954) > 0 else ''
        df_record = pd.DataFrame(data={'001': [mmsid], '954$a': [a_954], '954$b': [b_954], '954$c': [c_954],
                                       '954$d': [d_954], '954$e': [e_954], '954$f': [f_954], '954$g': [g_954],
                                       '954$h': [h_954], '954$i': [i_954]})
        records.append(df_record)
    return pd.concat(records)       
```

#### Extract item information from MARC file
Note: this takes several minutes to run.

```python
with open('TC-p-enum-chron-20201110_new.mrc', 'rb') as f:
    records = extract_marc_info(f)
records
```

```python
#add libloc column
records['libloc'] = records['954$f'] + ' ' + records['954$g']
records
```

```python
#read in libloc spreadsheet
libloc = pd.read_excel('PCADlocationList2.xlsx', sheet_name='locationList')
libloc
```

```python
#add libloc column to spreadsheet df
libloc['libloc'] = libloc['Library Code'] + ' ' + libloc['Location Code']
libloc
```

```python
libloc_include = libloc[libloc['Include?'] == 'Yes']
libloc_include
```

```python
keep_locs = [x for x in libloc_include['libloc']]
keep_locs
```

```python
#filter records on basis of libloc
chron = records[records['libloc'].isin(keep_locs)]
chron
```

```python
chron_wdn_filtered = chron[(chron['954$i']) != 'WDN']
chron_wdn_filtered
```

```python
items_filtered = chron_wdn_filtered.drop(columns=['libloc'])
items_filtered
```

```python
#change filename
items_filtered.to_pickle(f'items-enumchron-{today}.pkl')
```
